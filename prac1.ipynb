### 1. Program to compute factorial of a given number

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

# Get input from the user
num = int(input("Enter a non-negative integer to compute its factorial: "))

# Compute and display the factorial
result = factorial(num)
print(f"The factorial of {num} is {result}")
```

### 2. Program to display a multiplication table for a given number

```python
def display_table(n):
    print(f"Multiplication Table for {n}:")
    for i in range(1, 11):
        print(f"{n} x {i} = {n * i}")

# Get input from the user
num_table = int(input("Enter a number to display its multiplication table: "))

# Display the table
display_table(num_table)
```

### 3. Program to identify if a given number is odd or even

```python
def check_odd_even(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Get input from the user
num_check = int(input("Enter an integer to check if it's odd or even: "))

# Check and display the result
result_oe = check_odd_even(num_check)
print(f"The number {num_check} is {result_oe}.")
```

### 4. Program to identify if the entered number is palindrome or not

```python
def is_palindrome(n):
    # Convert the number to a string for easy reversal
    str_num = str(n)
    # Compare the string with its reverse
    if str_num == str_num[::-1]:
        return True
    else:
        return False

# Get input from the user
num_pal = int(input("Enter a number to check if it's a palindrome: "))

# Check and display the result
if is_palindrome(num_pal):
    print(f"The number {num_pal} is a palindrome.")
else:
    print(f"The number {num_pal} is not a palindrome.")
```

### 5. Develop an Expert System that provides simple decision-making

Here's a basic example of a rule-based expert system for diagnosing a simple car problem. This system takes user input about car symptoms and provides a possible diagnosis.

```python
def car_expert_system():
    print("Welcome to the Simple Car Problem Expert System!")
    print("Please answer the following questions with 'yes' or 'no'.")

    symptoms = {
        "engine_cranks_but_wont_start": False,
        "no_power_at_all": False,
        "lights_are_dim": False,
        "engine_makes_clicking_sound": False,
        "fuel_gauge_is_empty": False
    }

    symptoms["engine_cranks_but_wont_start"] = input("Does the engine crank but not start? (yes/no): ").lower() == 'yes'
    symptoms["no_power_at_all"] = input("Is there no power at all (no lights, no horn)? (yes/no): ").lower() == 'yes'

    if symptoms["no_power_at_all"]:
        symptoms["lights_are_dim"] = input("Are the lights dim or not working? (yes/no): ").lower() == 'yes'
        symptoms["engine_makes_clicking_sound"] = input("Does the engine make a clicking sound when you try to start it? (yes/no): ").lower() == 'yes'

    symptoms["fuel_gauge_is_empty"] = input("Is the fuel gauge empty or very low? (yes/no): ").lower() == 'yes'

    # Decision-making rules
    if symptoms["fuel_gauge_is_empty"]:
        print("\nDiagnosis: Your car might be out of fuel. Try adding gas.")
    elif symptoms["no_power_at_all"] and (symptoms["lights_are_dim"] or symptoms["engine_makes_clicking_sound"]):
        print("\nDiagnosis: It sounds like a dead or very low battery. Check battery terminals or try a jump start.")
    elif symptoms["engine_cranks_but_wont_start"] and not symptoms["fuel_gauge_is_empty"]:
        print("\nDiagnosis: This could be a fuel pump issue, spark plug problem, or a sensor failure. You might need professional help.")
    elif symptoms["no_power_at_all"]:
        print("\nDiagnosis: This is likely a completely dead battery or a major electrical failure. Check battery and main fuses.")
    else:
        print("\nDiagnosis: I'm not sure based on the information provided. You might want to consult a mechanic.")

# Run the expert system
car_expert_system()
```
